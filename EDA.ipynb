```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns

# Handle missing values
df['text'].fillna('', inplace=True)

# Function to get the top N-grams
def get_top_ngrams(corpus, n=2, k=10):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)[:k]
    return words_freq

```

```python
# Get the top N-grams for the entire dataset
top_ngrams = get_top_ngrams(df['text'], n=2, k=10)

# Visualize the top N-grams
plt.figure(figsize=(10, 6))
sns.barplot(x=[x[0] for x in top_ngrams], y=[x[1] for x in top_ngrams])
plt.title('Top 10 Bi-grams in the Dataset')
plt.xlabel('Bi-gram')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Get the top N-grams for the entire dataset
top_ngrams = get_top_ngrams(df['text'], n=3, k=10)

# Visualize the top N-grams
plt.figure(figsize=(10, 6))
sns.barplot(x=[x[0] for x in top_ngrams], y=[x[1] for x in top_ngrams])
plt.title('Top 10 tri-grams in the Dataset')
plt.xlabel('tri-gram')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```
